# Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line

# Write a Python program to find the volume of a sphere with diameter 12 cm.  Formula: V=4/3 * π * r 3 

```python
diameter = 12
radius = diameter/2
pi = 3.1415926535897931
volume = (4/3) * pi * (radius**3)
print(volume)
```

# Create the below pattern using nested for loop in Python. 
 
 
*  
* *
* * *
* * * *
* * * * *
* * * *
* * *
* *
*  

```python
def trikone(li):
    for i in li:
        count = 0
        print('')
        while (count < i):
            print("*", end="")
            count = count + 1


list = range(0,6)
revlist = range(4,0,-1) 

# prints two triangle patterns of stars

trikone(list)
trikone(revlist)
```

# Write a Python Program to print the given string in the format specified in the ​sample output. 
 WE, THE PEOPLE OF INDIA, having solemnly resolved to constitute India into a SOVEREIGN, SOCIALIST, SECULAR, DEMOCRATIC REPUBLIC and to secure to all its citizens 
 
Sample Output: 
 
WE, THE PEOPLE OF INDIA,   having solemnly resolved to constitute India into a SOVEREIGN, !  SOCIALIST, SECULAR, DEMOCRATIC REPUBLIC    and to secure to all its citizens 

```python
def pushspace(n):
    count = 0
    while(count < n):
        print(" ",end="")
        count = count + 1
    #print('\n')

text = "WE, THE PEOPLE OF INDIA, having solemnly resolved to constitute India into a SOVEREIGN, ! SOCIALIST, SECULAR, DEMOCRATIC REPUBLIC and to secure to all its citizens"
pluck1=text.index('having')
pluck2=text.index('!')+1
pluck3=text.index('and')
pluck4=text.index('citizens')
l = len('citizens')
spacing = [5,10,13]

# WE, THE PEOPLE OF INDIA,
for i in range(0,pluck1) :
    print(text[i], end="")
print()

# having solemnly resolved to constitute India into a SOVEREIGN, ! 
pushspace(spacing[0])
for i in range(pluck1,pluck2) :
    print(text[i], end="")
print()

# SOCIALIST, SECULAR, DEMOCRATIC REPUBLIC 
pushspace(spacing[1])
for i in range(pluck2,pluck3) :
    print(text[i], end="")
print()

#   and to secure to all its citizens
pushspace(spacing[2])
for i in range(pluck3,pluck4+l) :
    print(text[i], end="")
```
